## Option 2: adata_stock.info.all_code() works. This command takes approx 5 min.

```python
from adata import stock as adata_stock
all_stocks_adata = adata_stock.info.all_code()
all_stocks_adata.to_csv("./all_stocks_adata.csv", index=False, encoding="utf-8-sig")
all_stocks_adata.head()
```

## Option 3: akshare.stock_info_a_code_name() works. This command takes approx 2 min.

```python
import akshare as ak
all_stocks_akshare = ak.stock_info_a_code_name()
all_stocks_akshare.to_csv("./all_stocks.csv", index=False, encoding="utf-8-sig")
all_stocks_akshare.head()
```
